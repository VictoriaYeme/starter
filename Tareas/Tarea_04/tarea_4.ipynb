```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define your data
data = {
    "Día": ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado"],
    "Duración": [1, 1.30, 0, 1.30, 0, 1],
}

df = pd.DataFrame(data)

# Assign numeric values to days (just an example mapping)
day_mapping = {"Lunes": 1, "Martes": 2, "Miércoles": 3, "Jueves": 4, "Viernes": 5, "Sábado": 6}
df["Día_Num"] = df["Día"].map(day_mapping)

sns.set_theme(style="whitegrid")

# Plot the line graph
sns.lineplot(data=df, x="Día", y="Duración", linewidth=2.5)

# Set the x-axis labels to be the actual days
plt.xticks(list(day_mapping.values()), list(day_mapping.keys()))

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define your own data
data = {
    "Día": ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado"],
    "Duración Total": [60, 90, 0, 90, 0, 60],  # Duración en minutos
    "Calentamiento": [10, 10, 0, 10, 0, 10],  # Duración en minutos
    "Clases Técnicas": [20, 20, 0, 20, 0, 20],  # Duración en minutos
    "Ensayo Coreografía": [30, 60, 0, 60, 0, 30],  # Duración en minutos
}

df = pd.DataFrame(data)

# Configure Seaborn style
sns.set_theme(style="darkgrid")

# Create a distribution plot
g = sns.displot(
    df, x="Duración Total", col="Día",
    binwidth=10, height=3, facet_kws=dict(margin_titles=True),
)

# Customize the plot
g.set_axis_labels("Duración (minutos)", "Cantidad")
g.set_titles(col_template="{col_name}")
g.fig.suptitle("Distribución de Duración por Día")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define your own data
data = {
    "Día": ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado"],
    "Duración Total": [60, 90, 0, 90, 0, 60],  # Duración en minutos
    "Calentamiento": [10, 10, 0, 10, 0, 10],  # Duración en minutos
    "Clases Técnicas": [20, 20, 0, 20, 0, 20],  # Duración en minutos
    "Ensayo Coreografía": [30, 60, 0, 60, 0, 30],  # Duración en minutos
}

df = pd.DataFrame(data)

# Pivot the data to create a format suitable for a heatmap
df_heatmap = df.pivot(index="Día", columns="Duración Total", values=["Calentamiento", "Clases Técnicas", "Ensayo Coreografía"])

# Configure Seaborn style
sns.set_theme()

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(df_heatmap, annot=True, fmt=".0f", linewidths=.5, ax=ax)  # Use fmt=".0f" for float values

# Customize the plot
plt.title("Heatmap de Duración por Categoría por Día")
plt.show()
```
